```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
```

```python
titanic = pd.read_csv('/content/train (1).csv')
```

```python
flights = sns.load_dataset('flights')
```

```python
iris = sns.load_dataset('iris')
```

# **ScatterPlot (numerical-numerical)**

```python
tips
```

```python
sns.scatterplot(x = tips['total_bill'],y = tips['tip'])
plt.show()
```

```python
tips['total_bill'].corr(tips['tip'])
```

```python
#multivariate here -> 5 infos here

sns.scatterplot(x = tips['total_bill'],y = tips['tip'], hue = tips['sex'],style=tips['smoker'],size=tips['size'])
plt.show()
```

# **BarPlot (Numerical-Categorical)**

```python
titanic.head()
```

```python
titanic.groupby('Pclass')['Age'].max()
```

```python
sns.barplot(x = 'Pclass',y = 'Age',data = titanic)
plt.show()
```

```python
sns.barplot(
    x='Pclass',
    y='Age',
    data=titanic,
    estimator=max
)
plt.show()
```

```python
sns.barplot(x = 'Pclass', y = 'Fare', data=titanic, hue=titanic['Sex'])
```

# **BoxPlot(Numerical-Cat)**

```python
sns.boxplot(x=titanic['Sex'], y=titanic['Age'], data=titanic,hue=titanic['Survived'])
```

# **DistPlot**

```python
sns.distplot(titanic['Age'])
```

```python
titanic.groupby(titanic['Survived'])['Age']
```

```python
sns.distplot(titanic[titanic['Survived'] == 0]['Age'],hist=False)
sns.distplot(titanic[titanic['Survived'] == 1]['Age'],hist=False)
plt.show()
```

```python
sns.distplot(titanic[titanic['Survived'] == 0]['Age'])
plt.show()
```

# **HeatMap(Cat-Cat)**

```python
titanic.head()
```

```python
pd.crosstab(titanic['Pclass'], titanic['Survived'])
```

```python
sns.heatmap(pd.crosstab(titanic['Pclass'], titanic['Survived']))
```

```python
titanic.groupby('Survived')['Pclass'].mean()
```

```python
(titanic.groupby('Pclass')['Survived'].mean()*100).plot(kind='bar')
```

```python
# To count the occurrences of each sex within each Pclass:
titanic.groupby('Pclass')['Sex'].value_counts()
```

```python
titanic.groupby('Sex')['Survived'].mean()
```
