```python
all_dataframes = []
for i in range(1,10):
  response = requests.get(f"https://api.themoviedb.org/3/discover/movie?include_adult=false&include_video=false&language=en-US&page={i}&sort_by=popularity.desc", headers=headers)
  temp_df = pd.DataFrame(response.json()['results'])[['id','adult','original_title','overview','release_date']]
  all_dataframes.append(temp_df)

df = pd.concat(all_dataframes, ignore_index=True)
```
