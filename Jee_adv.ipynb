```python
##QQ plot
import scipy.stats as stats
# Ensure the 'Combined' column is numeric and handle any non-numeric entries
data = pd.to_numeric(df['diff'], errors='coerce').dropna()
stats.probplot(data, dist = 'norm', plot = plt)
plt.show()
```

```python
print(df['diff'].mean())
print(df['diff'].median())
print(df['diff'].mode())
print(df['diff'].std())
print(df['diff'].skew())
```

```python
from scipy.stats import shapiro

stats,p = shapiro(df['diff'])
print(stats,p)
```

```python
mean_diff = df['diff'].mean()
median_diff = df['diff'].median()
mode_diff = df['diff'].mode()
std_diff = df['diff'].std()
skew_diff = df['diff'].skew()
```

```python
#3sigma limit

upper_3sigma = mean_diff + 3*std_diff
lower_3sigma = mean_diff - 3*std_diff

print(upper_3sigma)
print(lower_3sigma)
```

```python
df[(df['diff'] >=  upper_3sigma) | (df['diff'] <= lower_3sigma)].drop(columns = ['combined_scale','diff'])
```

```python
#4sigma limit

upper_4sigma = mean_diff + 4*std_diff
lower_4sigma = mean_diff - 4*std_diff

print(upper_4sigma)
print(lower_4sigma)

df[(df['diff'] >=  upper_4sigma) | (df['diff'] <= lower_4sigma)].drop(columns = ['combined_scale','diff'])
```

```python
#5sigma limit

upper_5sigma = mean_diff + 5*std_diff
lower_5sigma = mean_diff - 5*std_diff

print(upper_5sigma)
print(lower_5sigma)

df[(df['diff'] >=  upper_5sigma) | (df['diff'] <= lower_5sigma)].drop(columns = ['combined_scale','diff'])
```

```python
plt.plot(df[(df['diff'] >=  upper_5sigma) | (df['diff'] <= lower_5sigma)].drop(columns = ['combined_scale','diff'])['abs_diff'])
```

```python
df[(df['diff'] >=  upper_5sigma) | (df['diff'] <= lower_5sigma)].drop(columns = ['combined_scale','diff']).sort_values(by = ['abs_diff'],ascending = False)
```
