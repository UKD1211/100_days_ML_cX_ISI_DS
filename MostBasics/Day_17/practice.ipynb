```python
df['title'].str.split(",")
```

```python
def lol(l:list):
  return l[0]
print(lol([1,2,3]))
```

```python
df['JobDesignation'] = df['title'].str.split(",").apply(lol)
```

```python
df['JobDesignation'].value_counts()
```

```python
import matplotlib.pyplot as plt
```

```python
job_counts = df['JobDesignation'].value_counts()
plt.bar(job_counts.index, job_counts.values)
```

```python
import matplotlib.pyplot as plt

job_counts = df['JobDesignation'].value_counts()

plt.figure(figsize=(12, 6))                  # make the plot bigger
plt.bar(job_counts.index, job_counts.values)
plt.xticks(rotation=90, ha='right')          # rotate labels so they dont overlap
plt.xlabel("Job Designation")
plt.ylabel("Count")
plt.tight_layout()                           # adjust layout so labels fit
plt.show()
```
