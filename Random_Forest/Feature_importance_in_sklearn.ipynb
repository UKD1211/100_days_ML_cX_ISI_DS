```python
import numpy as np
import pandas as pd
```

```python
from sklearn.datasets import fetch_openml
```

```python
mnist = fetch_openml('mnist_784', version=1)

X = mnist.data      # shape (70000, 784)
y = mnist.target    # labels 0-9

print(X.shape)
print(y.shape)
```

```python
X
```

```python
import seaborn as sns
sns.heatmap(X.iloc[7009].values.reshape(28,28))
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf  = RandomForestClassifier()
rf.fit(X, y)
```

```python
arr = rf.feature_importances_
```

```python
arr
```

```python
sns.heatmap(arr.reshape(28,28))
```

Feature Importance in DT

```python
from sklearn.datasets import make_classification
```

```python
X,y = make_classification(n_samples = 50, n_features = 2, n_redundant= 0 , n_informative= 2 , random_state = 42, n_classes= 2)
```

```python
sns.scatterplot(x = X[:,0], y = X[:,1], c = y)
```

```python
from sklearn.tree import DecisionTreeClassifier,plot_tree
```

```python
clf = DecisionTreeClassifier()
clf.fit(X,y)
```

```python
plot_tree(clf)
```

```python
clf.feature_importances_
```

```python
  from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier(n_estimators = 2)
```

```python
rf.fit(X,y)
```

```python
rf.feature_importances_
```
