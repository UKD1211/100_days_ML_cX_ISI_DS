```python
mean_yes = yes.mean()
mean_no = no.mean()

std_yes = yes.std()
std_no = no.std()

n_yes = len(yes)
n_no = len(no)

pooled_std = np.sqrt(((n_yes-1)*std_yes**2 + (n_no-1)*std_no**2) / (n_yes+n_no-2))

cohens_d = (mean_yes - mean_no) / pooled_std
cohens_d
```
