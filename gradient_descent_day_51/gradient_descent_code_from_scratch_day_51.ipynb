```python
class GDRegressor:
  def __init__(self,learning_rate,epochs):
    self.m = 100
    self.b = -120
    self.lrate = learning_rate
    self.epochs = epochs

  def fit(self,X,y):
        # calcualte the b using GD
        for i in range(self.epochs):
            loss_slope_b = -2 * np.sum(y - self.m*X.ravel() - self.b)
            loss_slope_m = -2 * np.sum((y - self.m*X.ravel() - self.b)*X.ravel())

            self.b = self.b - (self.lr * loss_slope_b)
            self.m = self.m - (self.lr * loss_slope_m)
        print(self.m,self.b)

  def predict(self,X):
        return self.m * X + self.b
```

```python
class uttamGD:
  def __init__(self,learning_rate = 0.01,epochs = 20):
    self.m = -100
    self.b = -120
    self.lr = learning_rate
    self.epochs = epochs
  def fit(self,X_train,y_train):
    for i in range(self.epochs):
      m_slope = -2* np.sum((y_train - self.m*X_train.ravel() - self.b)*X_train.ravel())
      b_slope = -2 * np.sum(y_train - self.m * X_train.ravel() - self.b)
      self.m = self.m - self.lr*m_slope
      self.b = self.b - self.lr*b_slope
      y_pred = self.m*X_train + self.b
      plt.plot(X_train,y_pred)
    print(self.m)
    print(self.b)
    plt.plot(X_train,self.m*X_train + self.b,color = 'yellow')

  def predict(self,X_test):
    return self.m*X_test + self.b
```

```python
gd = uttamGD()
```

```python
gd.fit(X_train,y_train)
```

```python
y_pred_new = gd.predict(X_test)
```
