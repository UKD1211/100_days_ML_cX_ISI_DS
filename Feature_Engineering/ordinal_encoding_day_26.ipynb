```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("/content/customer.csv")
```

```python
df = df.iloc[:,2:]
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(df.drop("purchased",axis=1),df['purchased'],test_size = 0.2
,random_state=0)
```

```python
X_train.shape
```

```python
from sklearn.preprocessing import OrdinalEncoder
oe = OrdinalEncoder(categories=[['Poor','Average','Good'],['School','UG','PG']])
oe.fit(X_train)
```

```python
X_train = oe.transform(X_train)
X_test = oe.transform(X_test)
```

```python
#label encoding is only for the output columns explicitely , not the input x
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
le.fit(y_train)
```

```python
from sklearn.model_selection import train_test_split

# Re-run train_test_split to get fresh categorical y_train and y_test
# This ensures y_train_cat and y_test_cat are in their original string format ('No', 'Yes')
_, _, y_train_cat, y_test_cat = train_test_split(df.drop("purchased",axis=1), df['purchased'], test_size = 0.2, random_state=0)

# Now, transform these categorical series using the fitted LabelEncoder
y_train = le.transform(y_train_cat)
y_test = le.transform(y_test_cat)
```
