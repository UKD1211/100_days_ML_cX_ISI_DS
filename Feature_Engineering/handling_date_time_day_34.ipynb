```python
date['date'] = pd.to_datetime(date['date'])
```

```python
date
```

```python
date['date_year'] = date['date'].dt.year
```

```python
date.head()
```

```python
date['date_month'] = date['date'].dt.month_name()
```

```python
date
```

```python
date['day'] = date['date'].dt.day_name()
```

```python
date
```

```python
# is weekend?

date['date_is_weekend'] = np.where(date['day'].isin(['Sunday', 'Saturday']), 1,0)

date.drop(columns=['product_id','city_id','orders']).head()
```

```python
date['date_week'] = date['date'].dt.isocalendar().week
```

```python
date
```

```python
date['quarter'] = date['date'].dt.quarter

date.drop(columns=['product_id','city_id','orders']).head()
```

```python
date['semester'] = np.where(date['quarter'].isin([1,2]), 1, 2)

date.drop(columns=['product_id','city_id','orders']).head()
```

```python
#time elapsed between dates

import datetime as dt

date['time_elapsed'] = date['date'] - date['date'].min()

date
```

```python
import datetime
today = datetime.datetime.today()
today
```

```python
today - date['date']
```

```python
(today - date['date']).dt.days
```

```python
# Months passed

months_passed = (today.year - date['date'].dt.year) * 12 + (today.month - date['date'].dt.month)
print(months_passed)
```
