```python
import numpy as np

class UKDrgd:
    def __init__(self, learning_rate=0.01, epochs=1000, alpha=0.01):
        self.lr = learning_rate
        self.epochs = epochs
        self.alpha = alpha
        self.coef_ = None

    def fit(self, X_train, y_train):
        n_samples, n_features = X_train.shape

        # initialize weights
        self.coef_ = np.zeros(n_features)

        for _ in range(self.epochs):
            # prediction
            y_pred = np.dot(X_train, self.coef_)

            # gradient (Ridge)
            gradient = (1/n_samples) * np.dot(X_train.T, (y_pred - y_train)) \
                       + self.alpha * self.coef_

            # update
            self.coef_ = self.coef_ - self.lr * gradient

        print("Weights:", self.coef_)

    def predict(self, X_test):
        return np.dot(X_test, self.coef_)
```
